```python
data_books.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
data_books
```

```python
# Первинне дослідження даних
#Чи є в якихось змінних пропуски?
# Відповідь: Ні
data_books.isnull().sum()
```

```python
# Визначте, яка ціна у нас максимальна, мінімальна, середня, медіанна (використовуйте функції max, min, mean, median)
# Відповідь: Максимальна ціна? 105
# Відповідь: Мінімальна ціна?  0
# Відповідь: Середня ціна?     13.1
# Відповідь: Медіанна ціна?    11.0
price = data_books['price']
max_price = price.max()
min_price = price.min()
mean_price = price.mean()
median_price = price.median()
print(f'Max price: {max_price}\nMin price: {min_price}\nMean price: {mean_price}\nMedian price: {median_price}')
```

```python
#Пошук та сортування даних
# 1. Який рейтинг у датасеті найвищий? Відповідь:
rating = data_books['user_rating'].max()
rating
#Відповідь: 4.9
```

```python
#2. Скільки книг мають такий рейтинг?
lst = 0
for i in data_books['user_rating']:
    if i == 4.9:
        lst += 1
lst
#Відповідь: 52
```

```python
# 3. Яка книга має найбільше відгуків?
reviews = data_books[data_books['reviews'] == data_books['reviews'].max()]
print(reviews.name)
#Відповідь: Where the Crawdads Sing
```

```python
# 4. З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)?
top_2015 = data_books[data_books['year'] == 2015]
price_max = top_2015[top_2015['price'] == top_2015['price'].max()]
price_max.name
#Відповідь: Publication Manual of the American Psychologic...
```

```python
# 5. Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)?
# Another way
# top_2010 = data_books[data_books['year'] == 2010]
# fiction = top_2010[top_2010['genre'] == 'Fiction']
# fiction.shape
result_2010 = data_books[(data_books['year'] == 2010) & (data_books['genre'] == 'Fiction')]
result_2010.shape
#Відповідь: 20
```

```python
# 6. Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)?
# Another way
# top_2011 = data_books[data_books['year'] == 2011]
# all_top = pd.concat([top_2010, top_2011], axis = 0)
# result = all_top[all_top['user_rating'] == 4.9]
# result
result = data_books[(data_books['year'] == 2011) & (data_books['user_rating'] == 4.9) |
                    (data_books['year'] == 2010) & (data_books['user_rating'] == 4.9)]
result
#Відповідь: 1
```

```python
# 7. І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).
new_top_2015 = top_2015[top_2015['price'] < 8]
books_in_2015 = new_top_2015.sort_values(by='price', ascending = True)
books_in_2015
```

```python
# 8. Яка книга остання у відсортованому списку?
last_book = books_in_2015.tail(1)
last_book.name
#Відповідь: Old School (Diary of a Wimpy Kid #10)
```

```python
# 1. Максимальна ціна для жанру Fiction:
grouped_genre = data_books.groupby(['genre']).agg({'price': ['max', 'min']})                                  
grouped_genre
#Відповідь: 82
# 2. Мінімальна ціна для жанру Fiction:
#Відповідь: 0
# 3. Максимальна ціна для жанру Non Fiction:
#Відповідь: 105
# 4. Мінімальна ціна для жанру Non Fiction:
#Відповідь: 0
```

```python
# Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні
data_books = data_books.drop_duplicates(subset=["name"])
grouped_author = data_books.groupby(['author'], group_keys=False)["name"].count()
# 5. Якої розмірності вийшла таблиця?
new_grouped_author = pd.DataFrame(grouped_author)
print(new_grouped_author.shape)
#Відповідь: 248
# 6. Який автор має найбільше книг?
new_grouped_author.columns = ['books']
end_version_author = new_grouped_author.sort_values(by='books', ascending = True)
print(end_version_author.tail(1))
#Відповідь: Jeff Kinney
# 7. Скільки книг цього автора?
#Відповідь:12
```

```python
# 8. У якого автора середній рейтинг мінімальний?
grouped_rating = data_books.groupby(['author']).agg({'user_rating': 'mean'})
end_version_rating = grouped_rating.sort_values(by='user_rating', ascending=False)
end_version_rating.tail(1)
#Відповідь: Donna Tartt
# 9. Який у цього автора середній рейтинг?
#Відповідь: 3.9
```

```python
# З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

new_data_books = pd.concat([end_version_author, end_version_rating], axis=1)
last_sort = new_data_books.sort_values(by=['books', 'user_rating'], ascending=True)
last_sort
# 10. Який автор перший у списку?
#Відповідь: Donna Tartt
```
