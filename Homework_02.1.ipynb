```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Вивести перші рядки таблиці за допомогою методу head
tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
                   match = 'Коефіцієнт народжуваності в регіонах України')
table = tmp[0]
print(table)
```

```python
#1. Вивести перші рядки таблиці за допомогою методу head
task_1 = table.head()
print(f'First 5 headers:\n{task_1}')
```

```python
# 2. Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(f'Strings and columns: \n{table.shape}')
```

```python
# 3. Замініть у таблиці значення "—" на значення NaN
table = table.replace("—", np.NaN)
print(table)
```

```python
# 4. Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(f'Types all columns: \n{table.dtypes}')
```

```python
# 5. Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
table = pd.concat([table["Регіон"], table.iloc[:, 1:].astype(np.float64)], axis=1)
table
```

```python
# 6. Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
task_6 = table.isnull().sum()
print(task_6)
```

```python
# 7. Видаліть з таблиці дані по всій країні, останній рядок таблиці
table = table[:-1]
print(table)
```

```python
# 8. Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
tab = table.fillna(table.mean())
tab
```

```python
# 9. Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
tab_2019 = tab[tab['2019'] > tab['2019'].mean()][['Регіон', '2019']]
tab_2019
```

```python
# 10. У якому регіоні була найвища народжуваність у 2014 році?
tab_2014_max = tab[tab['2014'] == tab['2014'].max()][['Регіон', '2014']]
tab_2014_max
```

```python
# 11. Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
regions = tab['Регіон']
birth = tab['2019']
plt.figure(figsize=(14, 10))
plt.barh(regions, birth)
plt.show()
```

```python
regions = tab['Регіон']
birth = tab['2019']
plt.figure(figsize=(14, 10))
plt.pie(birth, labels=regions, autopct="%.2f%%", pctdistance=0.9)
plt.show()
```

```python
regions = tab_2019['Регіон']
plt.figure(figsize=(14, 8))
plt.plot(regions, tab_2019['2019'], label='Народжуванність у 2019')
plt.title('Народжуванність у містах вище за середню у 2019 році')
plt.legend()
plt.show()
```
