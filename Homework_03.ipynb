```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Напишіть функцію гіпотези лінійної регресії у векторному вигляді;
def h(x, w_0):
    return np.dot(x, w_0)
```

```python
# 2. Cтворіть функцію для обчислення функції втрат у векторному вигляді;
def loss_func(x, y, w_0):
    n = x.shape[0]
    return ((h(x, w_0) - y) ** 2).sum() / (2*n)
```

```python
# 3. Реалізуйте один крок градієнтного спуску;

def grad_step(w_0, grad, learning_rate = 0.001):
    w_0 = w_0 - learning_rate * grad
    return w_0


def grad(x, y, w_0):
    n = x.shape[0]
    result = np.dot(x.T, (h(x, w_0) - y))
    return result / n


def grad_descent(x, y, w_0, num_iter=1000, learning_rate = 0.001, epsilon = 1e-10):
    loss = loss_func(x, y, w_0)
    loss_history = [loss]
    
    for i in range(num_iter):
        d_w = grad(x, y, w_0)
        w_0 = grad_step(w_0, d_w, learning_rate=learning_rate)
        loss = loss_func(x, y, w_0)
        if abs(loss - loss_history[-1]) < epsilon:
            loss_history.append(loss)
            break
        
        loss_history.append(loss)
        
    return w_0, d_w, loss_history
```

```python
data_house = pd.read_csv('Housing.csv')
data_house
```

```python
def normalize(value):
  return (value - value.mean())/value.std()
```

```python
norm_data = pd.DataFrame()
norm_data['price'] = normalize(data_house.price)
norm_data['area'] = normalize(data_house.area)
norm_data['bedrooms'] = normalize(data_house.bedrooms)
norm_data['bathrooms'] = normalize(data_house.bathrooms)
norm_data
```

```python
y = norm_data["price"].values.reshape(-1, 1)
x = norm_data[['area', 'bathrooms', 'bedrooms']].values
N = x.shape[1]
w_0 = np.linspace(0, 0, N).reshape((N, 1))
```

```python
W, d_w, loss_history = grad_descent(x, y, w_0, 10000)
loss = loss_history[-1]
print(f'Values: {d_w}')
print(f'Loss func: {loss}')
```
